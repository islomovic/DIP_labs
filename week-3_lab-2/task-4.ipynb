# Task 4: Image Shearing

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("image.jpg")
h, w = image.shape[:2]
```

```python
mx = 0.67
shear_x = np.float32([[1, mx, 0],[0,1,0]])
sheared_x = cv2.warpAffine(image, shear_x, (w,h))

plt.imshow(cv2.cvtColor(sheared_x, cv2.COLOR_BGR2RGB))
plt.title(f"Shear X (m={mx})")
plt.axis("off")
plt.show()
```

```python
my = 0.5
shear_y = np.float32([[1,0,0],[my,1,0]])
sheared_y = cv2.warpAffine(image, shear_y, (w,h))

plt.imshow(cv2.cvtColor(sheared_y, cv2.COLOR_BGR2RGB))
plt.title(f"Shear Y (m={my})")
plt.axis("off")
plt.show()
```

## Discussion
Shearing shifts pixels proportional to the other axis, tilting the image and changing angles while preserving area.
